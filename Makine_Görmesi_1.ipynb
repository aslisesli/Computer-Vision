# GÖRÜNTÜ YUMUŞATMA

```python
res=cv2.imread("filter.jpg")
cekirdek=np.ones((5,5),np.float32)/25
```

```python
res2=cv2.filter2D(res,-1,cekirdek)
```

```python
resimler=[res,res2]
for i in range(2):
    plt.subplot(1,2,i+1),
    plt.imshow(resimler[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

# GAUSS YÖNTEMİ İLE BULANIKLAŞTIRMA

```python
res3=cv2.GaussianBlur(res,(5,5),0)
```

```python
resimler=[res,res3]
for i in range(2):
    plt.subplot(1,2,i+1),
    plt.imshow(resimler[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
